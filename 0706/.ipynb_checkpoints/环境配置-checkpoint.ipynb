# 环境配置

## 安装Anaconda

Anaconda指的是一个开源的Python发行版本，其包含了conda、Python等180多个科学包及其依赖项。因为包含了大量的科学包，Anaconda 的下载文件比较大（约 531 MB），如果只需要某些包，或者需要节省带宽或存储空间，也可以使用Miniconda这个较小的发行版（仅包含conda和 Python）。

我们可以将Anaconda想象成一个包含不同文件（spyder、jupyter notebook等）的文件夹。

下面是三种安装Anaconda的方法。

### 官网

[https://www.continuum.io/downloads]

注：需要用chrome打开

### 清华大学开源软件镜像站

[https://mirrors.tuna.tsinghua.edu.cn/anaconda/archive/]

首先，在tuna上选择合适的版本。下载过程中有加入环境变量和默认Anaconda自带的Python是默认版本两个选项。推荐选择第一个，第二个选不选都可以。

下载完毕后，我们在cmd中运行以下三条命令：

conda

python

ipython

如果显示结果如下图，则下载成功：



#### 激活环境
&emsp;&emsp;如果下载时勾选了加入环境变量，则可能弹出一条警告： This Python interpreter is in a conda environment...
可以到官网 [https://conda.io/activation] 寻找解决方法，简而言之就是两种方法：

1. 在cmd中输入conda activate myenv

&emsp;&emsp;这里的myenv应替换为自己的环境名或者路径，可输入conda env list 查看自己的环境名。
这是种一次性做法，也就是说，每次打开cmd 时都需要运行此命令。
2. 如果你还没激活环境，可以在Anaconda Prompt 中输入c:\Anaconda3\Scripts\activate base 来激活环境

&emsp;&emsp;这里的路径是自己保存Anaconda的路径，并不一定在C盘

#### 修改包管理镜像为国内源
tuna 官方指南[https://mirrors.tuna.tsinghua.edu.cn/help/anaconda/]

&emsp;&emsp;我们先在cmd里运行命令：conda config --set show_channels_urls yes，会弹出一个名为.condarc的文件（如果没有弹出可以在C盘用户文件夹下寻找，找到后用记事本打开），将其中内容修改为：

channels:
  - defaults
show_channel_urls: true
channel_alias: https://mirrors.tuna.tsinghua.edu.cn/anaconda
default_channels:
  - https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/main
  - https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/free
  - https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/r
  - https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/pro
  - https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/msys2
custom_channels:
  conda-forge: https://mirrors.tuna.tsinghua.edu.cn/anaconda/cloud
  msys2: https://mirrors.tuna.tsinghua.edu.cn/anaconda/cloud
  bioconda: https://mirrors.tuna.tsinghua.edu.cn/anaconda/cloud
  menpo: https://mirrors.tuna.tsinghua.edu.cn/anaconda/cloud
  pytorch: https://mirrors.tuna.tsinghua.edu.cn/anaconda/cloud
  simpleitk: https://mirrors.tuna.tsinghua.edu.cn/anaconda/cloud
  
  然后保存即可。最后运行conda clean -i 清楚索引缓存。

## conda 常用命令

在cmd 中运行一些简单的命令可以帮助我们快速地获得需要的信息。

以下是一些常用的conda 命令：

### 查询信息

conda --version  # 查看conda版本号

conda list  # 查看已经下载的包

conda env list  # 查看环境

### 更新

conda update conda  # 更新conda版本，也会更新其他包  

conda update anaconda  # 更新anaconda版本  

conda update --all  # 更新所有包  

conda install package_name  # 更新指定包

### 环境

conda create --name your_env_name  # 创建新环境

conda remove --name your_env_name --all  # 删除某个环境

conda activate myenv  # 激活环境

conda deactivate  # 关闭环境

### 获取帮助

conda --help  # 获取帮助

conda env -h  # 查看环境管理的全部命令帮助


## cmd命令窗口和python交互模式的区别

命令行模式是直接在命令行窗口中运行python的py文件，需要先用文本编辑器编写代码，再通过命令行模式运行。交互模式是执行一行就返回一行代码的结果。

首先编写一个py文件，然后打开cmd，输入“python 文件路径”就可以执行这个py文件。如下图输入的是“python 1.py”。这种就是命令行模式。  
如果要进入交互模式，依然是打开cmd窗口或者在之前的cmd窗口继续操作，在窗口中输入python就进入交互模式了。
